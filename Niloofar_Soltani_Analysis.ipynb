# **MSU dataset**

## General

#### Dataset



```
    "Sem.":"No meaning",
    "Legenda":"Description of the variables showed for each row. It indicates which group (Production Line, Production Unit, Pescara Plant) the row is referred to.",
    "Linea di produzione":"Name of the production line",
    "Materiale":"Code of the final product",
    "Versione di prod.":"No meaning",
    "Data prod.":"Day of production",
    "Turno":"Production shift (every working day is divided into 3 shifts: First shift 06:00-14:00; Second shift 14:00-22:00; Third shift 22:00-06)",
    "%TSR":"Percentage of good products vs theorical products the line could have produced",
    "% Scarto":"Percentage of final products discarded among the total produced (if the product does not satisfy specific standards, it is discarded",
    "% Sched. Util.":"Percentage of time the line has been producing in 24 hours",
    "% Rate Util.":"Percentace of minutes the line can be used during 24 hours (usually it's 100%)",
    "% CU":"No meaning",
    "Resa":"Good products divided by the time the line was producing. Expressed in Pz/min",
    "Pz / min":"Number of products the line can potentially produce in 1 minute by design",
    "PZ Buoni in UMA":"Number of good products",
    "UMA":"Units (Unità di misura). Always MSU. MSU = 1000 SU. SU = Standrd Units. The standard unit is used by Fater to normalized every product so that different product quantities can be summed together",
    "PZ Buoni":"Number of good products",
    "PZ Totali":"Number of total products",
    "PZ Scarto":"Number of discarded products",
    "TN (min)":"Number of minutes the production line was on",
    "PZ/CAR":"No meaning",
    "PZ Totali in UMA":"Number of total products (good and discarded) expressed in MSU",
    "PZ Scarto in UMA":"Number of discarded products expressed MSU"
```

````python
# @title
import plotly.express as px
import pandas as pd

# Convert 'production_date' to datetime if it's not already
df_msu['production_date'] = pd.to_datetime(df_msu['production_date'])

# Define the range of years
years = range(2016, 2024)

# Loop through each year and generate plots
for year in years:
    # Filter the rows for the given year
    df_year = df_msu[df_msu['production_date'].dt.year == year]

    # Extract the month and day information
    df_year['Month'] = df_year['production_date'].dt.month
    df_year['Day'] = df_year['production_date'].dt.day

    # Count the occurrences of each month and day
    month_counts = df_year['Month'].value_counts().sort_index()
    day_counts = df_year['Day'].value_counts().sort_index()

    # Create interactive plot for months
    fig_month = px.bar(month_counts, x=month_counts.index, y=month_counts.values,
                       labels={'x': 'Month', 'y': 'Frequency'},
                       title=f'Frequency of Dates by Month in {year}')

    # Add hover information
    fig_month.update_traces(text=month_counts.values, textposition='outside', hoverinfo='x+y')

    # Add a description at the bottom
    fig_month.add_annotation(
        text=f"Total bars: {len(month_counts)}",
        xref="paper", yref="paper", x=0.5, y=-0.2,
        showarrow=False, font=dict(size=12), align="center"
    )

    # Create interactive plot for days
    fig_day = px.bar(day_counts, x=day_counts.index, y=day_counts.values,
                     labels={'x': 'Day', 'y': 'Frequency'},
                     title=f'Frequency of Dates by Day in {year}')

    # Add hover information
    fig_day.update_traces(text=day_counts.values, textposition='outside', hoverinfo='x+y')

    # Add a description at the bottom
    fig_day.add_annotation(
        text=f"Total bars: {len(day_counts)}",
        xref="paper", yref="paper", x=0.5, y=-0.2,
        showarrow=False, font=dict(size=12), align="center"
    )

    # Show plots
    fig_month.show()
    fig_day.show()
````

X-axis: Months (1 = January, 12 = December).

Y-axis: Frequency (number of times that month appears in production_date).

Bars: Show how often production occurred in each month.



---
If some months have much higher frequencies, production was more active during those months.

If a month has little or no bars, production was low or stopped in that period.


If some days (e.g., the 1st or 15th) have higher frequencies, production may follow a cycle (e.g., biweekly production).

If weekends (e.g., 6th-7th, 13th-14th) have lower bars, production may be reduced on weekends.

If some dates (e.g., 29th, 30th, 31st) are missing in multiple years, it could be because of shorter months (February) or irregular scheduling.
